```python
from os import path, listdir
from mne import io, read_epochs
import hmp
import xarray as xr
import pandas as pd
import statsmodels.formula.api as smf
```

```python
# Sampling rate. Downsampling to 250Hz is less computationally intensive, but loses some (but minimal) detail compared to 500Hz.
sfreq = 250

data_path = path.join('C:/Users/annad/Documents/Thesis/Data/Stimlocked_fixed_index/') # Change to folder where data is located
participant_lim = 0
subj_names = []

# Extracting subject name based on the file name
for x in listdir(data_path):
    if participant_lim >= 23: # In case you only want a few participants
        break

    # Only listing files of Task 2
    if '_t2.fif' in x:
        subj_names.append(x.split('.')[0])
        participant_lim += 1

# Create a list of files that can be read from the folder
subj_files = [data_path+x+'.fif' for x in subj_names]
```

```python
#Turning verbose off for the documentation but it is recommended to leave it on as some output from MNE might be useful
hmp.utils.read_mne_data(subj_files, epoched=True, sfreq=sfreq, verbose=False, subj_idx=subj_names, rt_col='rt', 
                                     reference='average', lower_limit_RT=.2, offset_after_resp=0.1, low_pass=50).to_netcdf(data_path+'Data_t2_250Hz.nc')
```

```python
# When using multiple CPUs explicitely setting the context of the multiprocessing is needed. The following line works in most environment
# mp.set_start_method("spawn")

# If multiprocessing errors/warning still present, comment previous line and uncomment this one
# mp.set_start_method("fork")
cpus = 5

# Load epochs from folder
epoch_data = xr.open_dataset(data_path+'Data_t2_250Hz.nc')

# Get channel information from the first participant
info = io.read_info(subj_files[0], verbose=False)

print(epoch_data)
info

# The following code is just to illustrate channel positions
single_epoch = read_epochs(subj_files[0])
single_epoch.plot_sensors(show_names=True);
```

```python
# Transform data and set variables
hmp_data = hmp.utils.transform_data(epoch_data, apply_zscore='all', n_comp=10)
event_width = 40
```

```python
# Select data where the SOA is long
hmp_data_long_soa = hmp.utils.condition_selection(hmp_data, epoch_data, 'long', variable='condition')
model_init_long = hmp.models.hmp(hmp_data_long_soa, epoch_data, sfreq=epoch_data.sfreq, cpus=cpus, event_width=event_width)

# Estimate maximal model using fit
max_model_long = model_init_long.fit()
```

```python
# Select data where the SOA is short
hmp_data_short_soa = hmp.utils.condition_selection(hmp_data, epoch_data, 'short', variable='condition')
model_init_short = hmp.models.hmp(hmp_data_short_soa, epoch_data, sfreq=epoch_data.sfreq, cpus=cpus, event_width=event_width)

# Estimate maximal model using fit
max_model_short = model_init_short.fit()
```

```python
# save fit
hmp.utils.save_fit(max_model_long, data_path+'fit_t2_long_250_ew40.nc')
hmp.utils.save_fit(max_model_short, data_path+'fit_t2_short_250_ew40.nc')
```

```python
# Plot fit
hmp.visu.plot_topo_timecourse(epoch_data, max_model_long, info, model_init_long, as_time=True, max_time=800)
hmp.visu.plot_topo_timecourse(epoch_data, max_model_short, info, model_init_short, as_time=True, max_time=800)
```

```python
# Use backwards estimation to validate model LONG
backward_model = model_init_long.backward_estimation(max_fit=max_model_long)
hmp.visu.plot_topo_timecourse(epoch_data, backward_model, info, model_init_long)
```

```python
# Use LOOCV to validate generalisation
loocv_model = hmp.loocv.loocv(model_init_long, hmp_data_long_soa, backward_model, print_warning=False)

# Colored shaded lines represent individuals
hmp.visu.plot_loocv(loocv_model, pvals=True, test='sign', indiv=True, mean=True)
```

```python
# Get stage durations per trial
times_short = hmp.models.hmp.compute_times(model_init_short, max_model_short, duration=True, as_time=True, onset=True, add_rt=True)
times_long = hmp.models.hmp.compute_times(model_init_long, max_model_long, duration=True, as_time=True, onset=True, add_rt=True)

df_short = times_short.to_dataframe().drop(columns=['component']).reset_index(drop=True)
df_long = times_long.to_dataframe().drop(columns=['component']).reset_index(drop=True)

# Add column with stage numbers and condition
df_short['stage'] = (df_short.index % 4) + 1
df_long['stage'] = (df_long.index % 4) + 1
df_short["condition"] = "short"
df_long["condition"] = "long"

data = pd.concat([df_short, df_long], ignore_index=True)
data.columns = ['participant', 'trial', 'duration', 'stage', 'condition'] # Change column names for consistency
data['stage'] = data['stage'].astype('category') # To make sure the mixed model sees the values as a categories

print(data)
data.to_csv('event_durations_t2.csv', index=False)
```

```python
# Create Mixed Linear Model to statistically test stage durations
model = smf.mixedlm("duration ~ condition * stage", data, groups=data['participant'])
result = model.fit()

p_values = result.pvalues
print(result.summary())
print(result.pvalues)
```

```python
# Get event onset times per trial (used when cropping epochs to event width)
times_short = hmp.models.hmp.compute_times(model_init_short, max_model_short, as_time=True, onset=True)
times_long = hmp.models.hmp.compute_times(model_init_long, max_model_long, as_time=True, onset=True)

df_short = times_short.to_dataframe().reset_index(drop=True)
df_long = times_long.to_dataframe().reset_index(drop=True)

# Add column with event numbers and condition
df_short['event'] = (df_short.index % 3) + 1
df_long['event'] = (df_long.index % 3) + 1
df_short["condition"] = "short"
df_long["condition"] = "long"

# Change column names for consistency
df_long.columns = ['participant', 'trial', 'onset', 'event', 'condition']
df_short.columns = ['participant', 'trial', 'onset', 'event', 'condition']

df_long.to_csv('event_onset_long_t2.csv', index=False)
df_short.to_csv('event_onset_short_t2.csv', index=False)
```
